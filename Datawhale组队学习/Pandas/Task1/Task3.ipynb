|（或） , &（且） , ~（取反） 的组合来实现

````python
# 多种条件组合
# 选出复旦大学中体重超过 70kg的大四学生, 或者北大男生中体重超过 80kg 的非大四的学生
cdt11 = df.School == "Fudan University"
cdt12 = df.Grade == "Senior"
cdt13 = df.Weight > 70

cdt21 = df.School == "Peking University"
cdt22 = df.Grade != "Senior"
cdt23 = df.Weight > 80

df[(cdt11 & cdt12 & cdt13) | (cdt21 & cdt22 & cdt23)].head()
````

````python
df.dtypes
````

````python
df.select_dtypes("float").head()
````

````python
def condition(x):
    condition_1_1 = x.School == 'Fudan University'
    condition_1_2 = x.Grade == 'Senior'
    condition_1_3 = x.Weight > 70
    condition_1 = condition_1_1 & condition_1_2 & condition_1_3
    condition_2_1 = x.School == 'Peking University'
    condition_2_2 = x.Grade == 'Senior'
    condition_2_3 = x.Weight > 80
    condition_2 = condition_2_1 & (~condition_2_2) & condition_2_3
    result = condition_1 | condition_2
    return result
# df.loc[condition(df)]
df.loc[condition]
````
