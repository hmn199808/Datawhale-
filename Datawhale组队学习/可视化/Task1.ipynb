# 2. 了解figure的组成
Figure：顶层级，用来容纳所有绘图元素

Axes：容纳大量元素用来构造一幅幅子图，一个figure可以由一个或多个子图组成

Axis：axes的下属层级，用于处理所有和坐标轴，网格有关的元素

Tick：axis的下属层级，用来处理所有和刻度有关的元素

# 3. 弄清figure和axes

Matplotlib在figures上绘制数据图，每个数据可以包含一个或多个Axes（可以根据xy坐标指定点的区域）。<br/>
通俗理解即：一个axes就是一个复杂多样性的图像，一个figure可以包含多个axes，且每个axes可以单独设置title和label<br/>


```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y1)
ax2.plot(x, y2)

ax1.set_title("x_y1")
ax1.set_xlabel("lable_x1")
ax1.set_ylabel("lable_y1")

ax2.set_title("x_y2")
ax2.set_xlabel("lable_x2")
ax2.set_ylabel("lable_y2")
```
